The business problem that we are looking to solve is deciding where in New York City would be the best to build an In-N-Out location. The data that we are looking at is August for 2016, 2017, 2018, and 2019. This will provide us with a standard month across multiple years to account for discrepencies in any specific year. The first thing that we will look at is which station has the highest amount of exits so that we know where the most people are walking around. 

In order to get to this value we are grouping by each individual turnstile and taking the difference between what the meter said in the previous period and what it says now to get the amount of people who have left that turnstile in that timeframe. The next step is grouping by station and by year. The graph below shows the Midtown West would be the best choice to build a new restuarant, because the top 4 stations are all in the same area.
